```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics
from PIL import Image
import colorsys
from colr import color
```

```python
def randwalk(p, n):
  RWPath = list()
  RWPath.append(-1 if random.random()<p else 1)
  for i in range(1,n):
    ZNValue = -1 if random.random()<p else 1
    XNValue = RWPath[i-1] + ZNValue
    RWPath.append(XNValue)
  return RWPath
```

```python
p = 0.5
n = 1000

r = randwalk(p,n)
g = randwalk(p,n)
b = randwalk(p,n)
test = randwalk(p,n)

plt.figure(figsize=(20,10))
plt.plot(r, color='r', label="RED")
plt.plot(g, color='g', label="GRN")
plt.plot(b, color='b', label="BLU")
plt.plot(test, color='black', label="XYZ")
plt.xlabel('Steps')
plt.ylabel('Value')
plt.legend(loc="upper left")
plt.show()

red = int(np.corrcoef(r, test)[0,1]*255)
grn = int(np.corrcoef(g, test)[0,1]*255)
blu = int(np.corrcoef(b, test)[0,1]*255)
print(red,grn,blu)
print(color('Hello there.', fore=(red, grn, blu), back=(red, grn, blu)))
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(r, color='r', linewidth=7)
plt.plot(g, color='g', linewidth=7)
plt.plot(b, color='b', linewidth=7)
plt.plot(test, color='black', linewidth=7)
fig.set_facecolor('black')
fig.show()
```

```python
im = Image.new("RGB", (1000,1000))
radius = min(im.size)/2.0
cx, cy = im.size[0]/2, im.size[1]/2
pix = im.load()
 
for x in range(im.width):
    for y in range(im.height):
        rx = x - cx
        ry = y - cy
        s = (rx ** 2.0 + ry ** 2.0) ** 0.5 / radius
        if s <= 1.0:
            h = ((math.atan2(ry, rx) / math.pi) + 1.0) / 2.0
            rgb = colorsys.hsv_to_rgb(h, s, 1.0)
            pix[x,y] = tuple([int(round(c*255.0)) for c in rgb])
 
im
```
